```python
hist_aer = xr.open_dataset(path_DAMIP+'/hist-aer/r1i1p1f1/Amon/tas/gr/latest/tas_Amon_'+model+'_hist-aer_r1i1p1f1_gr_185001-202012.nc').tas
hist_GHG = xr.open_dataset(path_DAMIP+'/hist-GHG/r1i1p1f1/Amon/tas/gr/latest/tas_Amon_'+model+'_hist-GHG_r1i1p1f1_gr_185001-202012.nc').tas
hist_nat = xr.open_dataset(path_DAMIP+'/hist-nat/r1i1p1f1/Amon/tas/gr/latest/tas_Amon_'+model+'_hist-nat_r1i1p1f1_gr_185001-202012.nc').tas
historical =  xr.open_dataset(path_CMIP+'/historical/r1i1p1f1/Amon/tas/gr/latest/tas_Amon_'+model+'_historical_r1i1p1f1_gr_185001-201412.nc').tas
```

```python
hist_list = [hist_aer, hist_GHG, hist_nat, historical]
hist_labels = ['hist_aer', 'hist_GHG', 'hist_nat', 'historical']
ts_list = [None]*len(hist_list)
ts_local_list = [None]*len(hist_list)

for i, hist in enumerate(hist_list):
    ts_list[i] = u.spatial_average(hist[:24])
    ts_local_list[i] = hist[:24].sel(lat=45.97835, lon=6.92643, method='nearest')
```

```python
for i, ts in enumerate(ts_list):
    ts.plot(label=hist_labels[i])
    
plt.legend()
plt.title('GMST time series (annual mean) 1850-2014/20 IPSL-CM6A-LR (r1i1p1f1)')
# plt.savefig('img/GMST_ts_1850-2020_IPSL-CM6A-LR_r1i1p1f1.jpg')
```

```python
for i, ts in enumerate(ts_local_list):
    ts.plot(label=hist_labels[i])
    
plt.legend()
plt.title('Argentire glacier (nearest grid cell) time series (annual mean) 1850-2014/20 IPSL-CM6A-LR (r1i1p1f1)')
# plt.savefig('img/GMST_ts_local_1850-2020_IPSL-CM6A-LR_r1i1p1f1.jpg')
```
