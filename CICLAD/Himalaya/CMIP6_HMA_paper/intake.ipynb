# Check data with Intake
conda env: `phd_v3` (in `envs/phd`)

```python
import intake
import pandas as pd
```

```python
url = "/thredds/ipsl/glipsl/ESPRI-CMIP6-test-collection.json"
col = intake.open_esm_datastore(url)
col
```

```python
col.df.head()
```

```python
# col.unique(["experiment_id"])
```

```python
activities = ['CMIP', 'ScenarioMIP']
experiments = ['historical', 'ssp126', 'ssp245', 'ssp370', 'ssp585']
# variables = ['tas']
# tabels = ['Amon]
variables = ['tas', 'snc', 'pr']
tables = ['Amon', 'LImon']
# members = ['r1i1p1f1']

query = dict(
    activity_id=activities, 
    experiment_id=experiments, 
    variable_id=variables, 
    table_id=tables, 
#     member_id=members
)
```

```python
# cat = col.search(require_all_on=["source_id"], **query)
cat = col.search(**query)
cat
```

```python
cat.df.head()
```

```python
cat.df.groupby("source_id")[
    ["activity_id", "experiment_id", "variable_id", "member_id"]
].nunique()
```

## Keep only the models that have all variables and experiments

```python
cat.df = cat.df.groupby("source_id").filter(
    lambda x : pd.Series(variables).isin(x['variable_id']).all()
)

cat.df = cat.df.groupby("source_id").filter(
    lambda x : pd.Series(experiments).isin(x['experiment_id']).all()
)
```

```python
cat.df.groupby("source_id")[
    ["activity_id", "experiment_id", "variable_id", "member_id"]
].nunique()
```

## Get the first member

```python
grouped = cat.df.groupby(['activity_id', 'institution_id', 'source_id', 'experiment_id', 'table_id', 'grid_label'])
grouped.first()
```

```python
cat.df = grouped.first().reset_index()
cat
```
