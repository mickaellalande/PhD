```python
t2m_era5 = ds(project='ClimServ_ERA5', variable='t2m', period='200901-201812', grid='GLOBAL_025', frequency='hourly')

print 'explore("choices"):'
print t2m_era5.explore('choices')
print ' -- '
print 'Summary:'
summary(t2m_era5)
```

```python
u_era5 = ds(project='ClimServ_ERA5', variable='u10', period='200901-201812', grid='GLOBAL_025', frequency='hourly')

print 'explore("choices"):'
print u_era5.explore('choices')
print ' -- '
print 'Summary:'
summary(u_era5)
```

```python
v_era5 = ds(project='ClimServ_ERA5', variable='v10', period='200901-201812', grid='GLOBAL_025', frequency='hourly')

print 'explore("choices"):'
print v_era5.explore('choices')
print ' -- '
print 'Summary:'
summary(v_era5)
```

```python
def subregion(x):
    output=llbox(x, lonmin=-5, lonmax=10, latmin=40, latmax=52) # France domain
    return output
```

```python
reduced_t2m=subregion(t2m_era5)
```

```python
reduced_u=subregion(u_era5)
```

```python
reduced_v=subregion(v_era5)
```

```python
ave_reduced_t2m=time_average(reduced_t2m)
ave_reduced_u=time_average(reduced_u)
ave_reduced_v=time_average(reduced_v)
```

```python
# Seasonal mean, t2m
annual_mean_t2m = clim_average(subregion(t2m_era5), 'ANM')
# DJF, MAM, JJA and SON climatologies
djf_t2m = clim_average(subregion(t2m_era5), 'DJF')
mam_t2m = clim_average(subregion(t2m_era5), 'MAM')
jja_t2m = clim_average(subregion(t2m_era5), 'JJA')
son_t2m = clim_average(subregion(t2m_era5), 'SON')
```

```python
# Seasonal mean, u
annual_mean_u = clim_average(subregion(u_era5), 'ANM')
# DJF, MAM, JJA and SON climatologies
djf_u = clim_average(subregion(u_era5), 'DJF')
mam_u = clim_average(subregion(u_era5), 'MAM')
jja_u = clim_average(subregion(u_era5), 'JJA')
son_u = clim_average(subregion(u_era5), 'SON')
```

```python
# Seasonal mean, v
annual_mean_v = clim_average(subregion(v_era5), 'ANM')
# DJF, MAM, JJA and SON climatologies
djf_v = clim_average(subregion(v_era5), 'DJF')
mam_v = clim_average(subregion(v_era5), 'MAM')
jja_v = clim_average(subregion(v_era5), 'JJA')
son_v = clim_average(subregion(v_era5), 'SON')
```
